## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
import random

# Define function to generate outcomes from each Trial
def roll_two_dice():
    return random.randint(1, 6) + random.randint(1, 6)

# start successful outcomes at zero
def simulate_experiment(num_iterations):
    count_successful_outcomes = 0

# Calculate score to evaluate against desired event condition
    for _ in range(num_iterations):
        total_score = roll_two_dice()

# Create function to evaluate score against desired event condition
        if total_score > 7 or total_score % 2 == 0:
            count_successful_outcomes += 1

# Determine probability 
    probability = count_successful_outcomes / num_iterations
    return probability

# Run the experiment 1000 times
num_iterations = 1000
result = simulate_experiment(num_iterations)

print(f"Estimated probability: {result:.4f}")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}
# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
```

```python
import random

# Function to simulate drawing 5 balls with replacement using random.choices
def draw_balls():
    return random.choices(list(ball_box.values()), k=5)

# Initialize counts for the outcomes
outcome_1_count = 0  # 3 white and 2 red balls
outcome_2_count = 0  # All balls of the same color

# Number of experiments
num_experiments = 1000

# Run the experiment 1000 times
for _ in range(num_experiments):
    drawn_balls = draw_balls()

    # Check outcome 1: 3 white and 2 red balls
    if drawn_balls.count("White") == 3 and drawn_balls.count("Red") == 2:
        outcome_1_count += 1

    # Check outcome 2: All balls of the same color
    if all(color == drawn_balls[0] for color in drawn_balls):
        outcome_2_count += 1

# Calculate probabilities
probability_outcome_1 = outcome_1_count / num_experiments
probability_outcome_2 = outcome_2_count / num_experiments

# Print results
print(f"Probability of drawing 3 white and 2 red balls: {probability_outcome_1:.4f}")
print(f"Probability of drawing all balls of the same color: {probability_outcome_2:.4f}")
```
